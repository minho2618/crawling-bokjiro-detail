```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Chrome 옵션 설정 (헤드리스 모드 필요 시 주석 해제)
options = webdriver.ChromeOptions()
# options.add_argument('--headless')

# WebDriver 초기화 (웹드라이버 매니저 사용)
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# 대기 시간 설정 (최대 10초)
wait = WebDriverWait(driver, 10)
def get_text_safe(selector, by=By.CSS_SELECTOR):
    """요소가 로드될 때까지 대기 후 텍스트를 반환합니다."""
    try:
        element = wait.until(EC.presence_of_element_located((by, selector)))
        return element.text.strip()
    except Exception as e:
        return f"[Error] {e}"

def click_safe(selector, by=By.CSS_SELECTOR):
    """요소가 클릭 가능해질 때까지 대기 후 JavaScript로 클릭합니다."""
    try:
        element = wait.until(EC.element_to_be_clickable((by, selector)))
        # 화면에 안 보일 경우를 대비해 JS 클릭 사용
        driver.execute_script("arguments[0].click();", element)
        time.sleep(2) # 클릭 후 렌더링 딜레이 반영
    except Exception as e:
        print(f"[Click Error] {e}")

url = "https://www.bokjiro.go.kr/ssis-tbu/twataa/wlfareInfo/moveTWAT52011M.do?wlfareInfoId=WLF00001107&wlfareInfoReldBztpCd=01"
driver.get(url)

# 1. 정책명
policy_name = get_text_safe('#uuid-7o > div > div > div')
print("1. 정책명:", policy_name)

# 2. 정책설명
policy_desc = get_text_safe('#uuid-7r > div > div')
print("2. 정책설명:", policy_desc)

# 3. 담당부처
dept = get_text_safe('#uuid-7u > div > div')
print("3. 담당부처:", dept)

# 4. 기준연도
year = get_text_safe('#uuid-7x > div > div')
print("4. 기준연도:", year)

# 5. 문의처
contact = get_text_safe('#uuid-7z > div > div')
print("5. 문의처:", contact)

# 6. 지원주기
cycle = get_text_safe('#uuid-81 > div > div')
print("6. 지원주기:", cycle)

# 7. 제공유형
prov_type = get_text_safe('#uuid-83 > div > div')
print("7. 제공유형:", prov_type)

# 8. 지원대상
click_safe("/html/body/div[1]/div[1]/div/div[3]/div/div/div/div[3]/div/div/div/div[2]/div/div/div/div[3]/div/div/div/div[2]/div/div/div[1]/div/div[1]/div/div/div[2]/div/a", by=By.XPATH)
target = get_text_safe('#uuid-8f > div > div', by=By.CSS_SELECTOR)
print("\n8. 지원대상:\n", target)

# 9. 서비스 내용
click_safe("/html/body/div[1]/div[1]/div/div[3]/div/div/div/div[3]/div/div/div/div[2]/div/div/div/div[3]/div/div/div/div[2]/div/div/div[1]/div/div[2]/div/div/div[2]/div/a", by=By.XPATH)
service_content = get_text_safe('#uuid-8o > div > div', by=By.CSS_SELECTOR)
print("\n9. 서비스 내용:\n", service_content)

# 10. 신청방법
click_safe("/html/body/div[1]/div[1]/div/div[3]/div/div/div/div[3]/div/div/div/div[2]/div/div/div/div[3]/div/div/div/div[2]/div/div/div[1]/div/div[3]/div/div/div[2]/div/a", by=By.XPATH)
apply_method = get_text_safe('#uuid-8u > div > div', by=By.CSS_SELECTOR)
print("\n10. 신청방법:\n", apply_method)

# 11. 추가정보
click_safe("/html/body/div[1]/div[1]/div/div[3]/div/div/div/div[3]/div/div/div/div[2]/div/div/div/div[3]/div/div/div/div[2]/div/div/div[1]/div/div[4]/div/div/div[2]/div/a", by=By.XPATH)
additional_info = get_text_safe('#uuid-9p > div > div', by=By.CSS_SELECTOR)
print("\n11. 추가정보:\n", additional_info)
```

## 데이터 저장 (JSON)

```python
import json
import os
from datetime import datetime

# 수집한 데이터 딕셔너리로 묶기
data = {
    "정책명": policy_name,
    "정책설명": policy_desc,
    "담당부처": dept,
    "기준연도": year,
    "문의처": contact,
    "지원주기": cycle,
    "제공유형": prov_type,
    "지원대상": target,
    "서비스 내용": service_content,
    "신청방법": apply_method,
    "추가정보": additional_info
}

# output 폴더 생성
os.makedirs("output", exist_ok=True)
today_str = datetime.today().strftime('%Y-%m-%d')

with open(f'output/{policy_name}-{today_str}.json', "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print(f"성공적으로 {policy_name}-{today_str}.json 파일에 저장되었습니다!")
```
